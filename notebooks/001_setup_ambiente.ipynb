## Conexão com o MongoDB Atlas

```python
import os
import pandas as pd
from pymongo import MongoClient
from dotenv import load_dotenv

# Carrega as variáveis do arquivo .env (que deve estar na mesma pasta do notebook ou na raiz do projeto)
load_dotenv(override=True)

# Coletando as credenciais
DB_USER     = os.getenv('DB_USER')
DB_PASSWORD = os.getenv('DB_PASSWORD')
DB_HOST     = os.getenv('DB_HOST')
DB_NAME     = "VendasArCondicionado"

# Montando a URI do Atlas
uri = f"mongodb+srv://{DB_USER}:{DB_PASSWORD}@{DB_HOST}/?retryWrites=true&w=majority&appName=trabalho-arq-medalhao-databricks"

try:
    # Instanciando o client
    client = MongoClient(uri)
    db = client[DB_NAME]
    
    # Executando um 'ping' para confirmar que a conexão de rede e credenciais estão OK
    client.admin.command('ping')
    
    print(f"✅ Conectado ao MongoDB Atlas no host [{DB_HOST}] com sucesso!")
    print(f"📂 Database ativa: {DB_NAME}")
    
except Exception as e:
    print(f"❌ Erro ao conectar: {e}")
```
